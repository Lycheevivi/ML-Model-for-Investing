# Section 1: Data Processing

```python
import pandas as pd
from google.colab import files

# Upload the file
uploaded = files.upload()

# The file name is 'sp500_fundamental_ratio_data.csv'
file_name = 'sp500_fundamental_ratio_data.csv'
```

```python
# Read the CSV file
df = pd.read_csv(file_name, parse_dates=["Report Date"])

# Set the index to 'Ticker' and 'Report Date'
df = df.set_index(["Ticker", "Report Date"])

# Print the DataFrame
print(df)
```

```python
# Drop rows with missing values
df = df.dropna()
df
```

```python
# Mark a price increase as "True" if it rises by more than 5%
threshold = 0.05 # 5%\

# Create the "Buy Signal" target variable
df["Buy Signal"] = df["PeriodReturn"] >= threshold
df
```

```python
# Count the number of True values in the target column
num_true_values = df["Buy Signal"].sum()

# Print the result
print(f"Number of True values in the target column: {num_true_values}")
```

```python
column_names = df.columns.tolist()
print("\nList of column names:")
print(column_names)
```

```python
# A list of features
features = [
'Altman Z-Score',
'Piotroski F-Score',
'Gross Profit Margin',
'Operating Margin',
'Net Profit Margin',
'Return on Equity',
'Return on Assets',
'Return On Invested Capital',
'Free Cash Flow to Net Income',
'Current Ratio',
'Debt Ratio',
'Dividend Payout Ratio',
'Net Debt / EBITDA',
 ]

# Shift the value of the features by a single period
df[features] = df.groupby("Ticker")[features].shift(1)
```

```python
# Eliminate the NaN generated after completing the shift, and remove the first row for each ticker.
df = df.loc[df.groupby("Ticker").cumcount() > 0]
```

# Section 3: Build Model and Prediction

We used year 2020 as the splitting year where data before year 2020 were used for training, data in year 2020 were used for testing, and data after year 2020 were used for validation purposes. After train, test, validation splitting, there are 8,351 rows of data for training, 1,130 rows of data for validation, and 2,394 rows of data for testing.

```python
# Train, Validation, Test data split
split_date = 2020

df_train = df.loc[df.index.get_level_values("Report Date").year < split_date]
df_validation = df.loc[df.index.get_level_values("Report Date").year == split_date]
df_test = df.loc[df.index.get_level_values("Report Date").year > split_date]

df_train.columns = df_train.columns.str.replace('[^A-Za-z0-9_]+', '', regex=True)
df_test.columns = df_test.columns.str.replace('[^A-Za-z0-9_]+', '', regex=True)
df_validation.columns = df_validation.columns.str.replace('[^A-Za-z0-9_]+', '', regex=True)
```

```python
!pip install lightgbm
```

```python
# Revised feature name after removing white space and special characters
features = [
'AltmanZScore',
'PiotroskiFScore',
'GrossProfitMargin',
'OperatingMargin',
'NetProfitMargin',
'ReturnonEquity',
'ReturnonAssets',
'ReturnOnInvestedCapital',
'FreeCashFlowtoNetIncome',
'CurrentRatio',
'DebtRatio',
'DividendPayoutRatio',
'NetDebtEBITDA',
 ]

# Build LGBM Classifier model
from lightgbm import LGBMClassifier

# Define classifier
estimator = LGBMClassifier(
 is_unbalance=True,
 max_depth=4,
 num_leaves=8,
 min_child_samples=400,
 n_estimators=50,
 )

# Fit model on training data
estimator.fit(df_train[features], df_train['BuySignal'])
```

```python
# Use test data to make prediction on validation set
df_validation["buy"] = estimator.predict(df_validation[features])
df_validation
```

```python
!pip install scikit-plot
```

```python
# Performance of the Model on the Validation Set
from sklearn.metrics import classification_report

print(classification_report(df_validation["BuySignal"], df_validation["buy"]))
```

```python
# Fit the model to the test dataset.
df_test["buy"] = estimator.predict(df_test[features])
df_test
```

```python
# Model Performance on Testing Set
from sklearn.metrics import classification_report

print(classification_report(df_test["BuySignal"], df_test["buy"]))
```

```python
# Portfolio Performance

# Choose the stocks that the model recommended
df_buy = df_test.loc[df_test["buy"] == True][["PeriodReturn", "BuySignal", "buy"]]
df_buy
```

```python
df_results = (
 df_buy.reset_index()
 .groupby("Report Date")
 .agg({"Ticker": "count", "PeriodReturn": "mean"})
 )

df_results
```

```python
df_results.describe()
```

```python
# Sharpe Ratio

import numpy as np

def sharpe(s_return: pd.Series, annualize: int, rf: float = 0) -> float:
    """
    Calculate sharpe ratio
    :param s_return: pd.Series with return
    :param annualize: int periods to use for annualization (252 daily, 12 monthly, 4 quarterly)
    :param rf: float risk-free rate
    :return: float sharpe ratio
    """
    # (mean - rf) / std
    sharpe_ratio = (s_return.mean() - rf) / s_return.std()

    # annualize
    sharpe_ratio = sharpe_ratio * np.sqrt(annualize)

    return sharpe_ratio
```

```python
sharpe_ratio = sharpe(df_results["PeriodReturn"], annualize=4)
print(f"Sharpe ratio: {round(sharpe_ratio, 2)}")
```

```python
# Use the quartly return, and calculate the cumulative returns
df_results["return_quart_cumulative"] = (df_results["PeriodReturn"] + 1).cumprod() - 1
df_results
```
